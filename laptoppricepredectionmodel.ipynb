```python
# Convert 'Memory' to string type
df['Memory'] = df['Memory'].astype(str)

# Replace 'GB' with an empty string and 'TB' with '000'
df["Memory"] = df["Memory"].str.replace('GB', '', regex=True)
df["Memory"] = df["Memory"].str.replace('TB', '000', regex=True)

# Split 'Memory' column into 'first' and 'second' columns
new = df["Memory"].str.split("+", n=1, expand=True)
df["first"] = new[0].str.strip()  # Strip whitespace

# Check if there is a second part in the split
if len(new.columns) > 1:
    df["second"] = new[1].str.strip()
else:
    df["second"] = None

# Extract the numeric part from 'first' and 'second' columns
df["first"] = df["first"].str.extract(r'(\d+)')  # Extract numbers using regex
df["second"] = df["second"].str.extract(r'(\d+)')  # Extract numbers using regex

# Convert 'first' and 'second' columns to integers, handling NaN values
df["first"] = pd.to_numeric(df["first"], errors='coerce').fillna(0).astype(int)
df["second"] = pd.to_numeric(df["second"], errors='coerce').fillna(0).astype(int)

# Ensure that the Layer columns exist before performing calculations
if {'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid', 'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid', 'Layer2Flash_Storage'}.issubset(df.columns):
    # Calculate the new memory columns
    df["HDD"] = (df["first"] * df["Layer1HDD"] + df["second"] * df["Layer2HDD"])
    df["SSD"] = (df["first"] * df["Layer1SSD"] + df["second"] * df["Layer2SSD"])
    df["Hybrid"] = (df["first"] * df["Layer1Hybrid"] + df["second"] * df["Layer2Hybrid"])
    df["Flash_Storage"] = (df["first"] * df["Layer1Flash_Storage"] + df["second"] * df["Layer2Flash_Storage"])

    # Drop unnecessary columns
    df.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid', 'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid', 'Layer2Flash_Storage'], inplace=True)
else:
    print("One or more Layer columns are missing from the dataframe.")
```
